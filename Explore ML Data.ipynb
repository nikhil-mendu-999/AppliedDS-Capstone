# Explore correlation in the ML data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the same data which is used for the Space X Falcon 9 First Stage Landing Prediction Lab and modify it.
X = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DS0701EN-SkillsNetwork/api/dataset_part_3.csv')
data = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DS0701EN-SkillsNetwork/api/dataset_part_2.csv')
target = data['Class'].to_numpy()
X['Class'] = target
X['Date'] = data['Date']
X.head()
```

```python
# Calculate corration between data
X.corr()
```

```python
# Calculate correlation with target
X.corr()[["Class"]].style.background_gradient(cmap = "bwr", vmin = -1, vmax = 1, axis = None)
```

```python
# Select only features with higher correlation with target
new_data = X[['Block', 'ReusedCount', 'GridFins_False', 'GridFins_True', 'Legs_False', 'Legs_True', 'Class']]
new_data.corr()[["Class"]].style.background_gradient(cmap = "bwr", vmin = -1, vmax = 1, axis = None)
```

```python
# Find the first grid fins usage
first_fins = X.loc[X['GridFins_True']== 1.0]
first_fins.sort_values('Date').head(1)
```

```python
# find the first landing legs usage
first_legs = X.loc[X['Legs_True']== 1.0]
first_legs.sort_values('Date').head(1)
```
